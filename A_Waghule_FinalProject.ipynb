```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
def convert1(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 1:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
#To extract the director name from each row we can use this function
#This function will only detect the job name=Director and append it to the list
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
df.cast = df.cast.apply(convert3)
```

```python
df.genres = df.genres.apply(convert1)
```

```python
df.production_companies = df.production_companies.apply(convert1)
```

### Linear Regression 

```python
X = newdf2.drop(['revenue'], axis=1)
y = newdf2['revenue']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.33, random_state = 5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,Y_train)
y_prediction =  lr.predict(X_train)
y_prediction
```

```python
# importing r2_score module

from sklearn.metrics import r2_score

from sklearn.metrics import mean_squared_error

# predicting the accuracy score

score=r2_score(Y_train,y_prediction)

print('r2 score = ',score)
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn import tree
```

```python
clf_en2 = DecisionTreeRegressor(criterion='absolute_error', max_depth=3, random_state=0)
d = clf_en2.fit(X_train, Y_train)
y_pred_en2 = clf_en2.predict(X_train)
```

### Random Forest

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
```

```python
# Import tools needed for visualization
from sklearn.tree import export_graphviz
import pydot
```

```python
feature_list2 = list(X.columns) #for future use
```

```python
# Instantiate model with 100 decision trees
rf2 = RandomForestRegressor(n_estimators = 100, random_state = 42)
# Train the model on training data
c = rf2.fit(X_train, Y_train)
```

```python
# Get numerical feature importances
importances = list(rf2.feature_importances_)
# List of tuples with variable and importance
feature_importances = [(feature, round(importance, 2)) for feature, importance in zip(feature_list2, importances)]
# Sort the feature importances by most important first
feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)
# Print out the feature and importances 
[print('Variable: {:20} Importance: {}'.format(*pair)) for pair in feature_importances];
```

```python
# Use the forest's predict method on the train data
predictions2 = rf2.predict(X_train)
# Calculate the absolute errors
errors2 = abs(predictions2 - Y_train)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors2), 2))
```

```python
rf2.score(X_train, Y_train)
```

```python
scores3 = cross_val_score(c, X_train, Y_train, cv=5)
print("%0.2f cross validation score with a standard deviation of %0.2f" % (scores3.mean(), scores3.std()))
```

```python
# Pull out one tree from the forest
tre2 = rf2.estimators_[5]
# Export the image to a dot file
export_graphviz(tre2, out_file = 'tree2.dot', feature_names = feature_list2, rounded = True, precision = 1)
# Use dot file to create a graph
(graph, ) = pydot.graph_from_dot_file('tree2.dot')
```

```python
text_representation2 = tree.export_text(tre2)
print(text_representation2)
```
